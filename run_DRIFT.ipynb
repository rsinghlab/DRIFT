## Running DRIFT to obtain diffused spatial transcriptomics input

```python
file_path = "data/MERMPH_4.h5ad" #Give your data file location here

adata = sc.read_h5ad(file_path)
adata.var_names_make_unique()

diffused_adata = DRIFT(adata, t = 2, k = 3) #t is the hyperparameter for heat kernel diffusion and k is the number of neighbors in the spatial graph
```

diffused_adata.X now contains the diffused inputs and can be used as input to foundation models for further spatial transcriptomics analysis
