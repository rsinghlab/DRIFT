```python
import sys
sys.path.append("..")   # add parent dir so alignment_code is visible


import pandas as pd
import matplotlib.pyplot as plt
import scanpy as sc
from src import align
```

```python


target_embedding_path = f"./data/adata_028_A1_filtered_embedding_Normal.h5ad"
source_embedding_path =  f"./data/adata_028_C1_filtered_embedding_Normal.h5ad"
```

```python
import scanpy as sc
target_embedding_adata = sc.read_h5ad(target_embedding_path)
target_coordinates = target_embedding_adata.obsm['spatial']

source_embedding_adata = sc.read_h5ad(source_embedding_path)
source_coordinates = source_embedding_adata.obsm['spatial']
```

```python
tar_features = pd.DataFrame(target_embedding_adata.X.T)
src_features = pd.DataFrame(source_embedding_adata.X.T)
pca_comb_features = align.get_pca_by_fit(tar_features, src_features)
pca_rgb_comb, pca_hex_comb = align.pca_to_hex_colors(pca_comb_features)

```

```python

plt.figure(figsize=(4,4))
plt.scatter(target_coordinates[:,0], target_coordinates[:,1], marker='o', s=5, c=pca_rgb_comb[:len(tar_features.T)])
plt.axis('off')
plt.title('Target - 28A1')
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.scatter(source_coordinates[:,0], source_coordinates[:,1], marker='o', s=5, c=pca_rgb_comb[len(tar_features.T):])
plt.axis('off')
plt.title('Source - 28C1')
plt.show()
```

```python
new_non_rigid_coordinates, new_rigid_coordinates = align.align_tissue(target_coordinates,                         
                                                                      source_coordinates, pca_comb_features)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Split PCA colors
pca_rgb_tar = pca_rgb_comb[:len(tar_features.T)]
pca_rgb_src = pca_rgb_comb[len(tar_features.T):]

# VERY tiny separation
z_tar = np.zeros(len(target_coordinates))
z_src = np.ones(len(new_rigid_coordinates)) * 0.02   # << super thin gap

fig = plt.figure(figsize=(8, 7))
ax = fig.add_subplot(111, projection='3d')

# Target (bottom)
ax.scatter(
    target_coordinates[:,0],
    target_coordinates[:,1],
    z_tar,
    c=pca_rgb_tar,
    s=8,
    alpha=0.7,
    label="Target"
)

# Aligned source (top)
ax.scatter(
    new_rigid_coordinates[:,0],
    new_rigid_coordinates[:,1],
    z_src,
    c=pca_rgb_src,
    s=8,
    alpha=0.85,
    label="Aligned (Rigid)"
)

# ---------------------------------------------------------
#  KEY FIX: Force the z-axis to zoom into the tiny gap 
# ---------------------------------------------------------
ax.set_zlim(0, 0.003)  # shrink z-axis range
ax.set_box_aspect([1, 1, 0.05])  # flatten the z dimension

ax.set_title("3D Overlay: Target + Aligned Source (Super Close Layers)")
ax.set_xticks([]); ax.set_yticks([]); ax.set_zticks([])
ax.legend(frameon=False)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Pick the label column here:
label_key = "ground_truth"   # <-- CHANGE THIS to the actual key

labels_tar = target_embedding_adata.obs[label_key].astype(str).values
labels_src = source_embedding_adata.obs[label_key].astype(str).values

# Build a consistent colormap
unique_labels = np.unique(np.concatenate([labels_tar, labels_src]))
cmap = plt.get_cmap("tab20")  # or tab10/tab20/tab20b
label_to_color = {lab: cmap(i % 20) for i, lab in enumerate(unique_labels)}

colors_tar = [label_to_color[l] for l in labels_tar]
colors_src = [label_to_color[l] for l in labels_src]
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Small separation so they appear on top of each other
z_tar = np.zeros(len(target_coordinates))
z_src = np.ones(len(new_rigid_coordinates)) * 0.2

fig = plt.figure(figsize=(8,7))
ax = fig.add_subplot(111, projection='3d')

# Target ground-truth
ax.scatter(
    target_coordinates[:,0],
    target_coordinates[:,1],
    z_tar,
    c=colors_tar,
    s=8,
    alpha=0.3,
    label="Target"
)

# Aligned source ground-truth
ax.scatter(
    new_rigid_coordinates[:,0],
    new_rigid_coordinates[:,1],
    z_src,
    c=colors_src,
    s=8,
    alpha=0.3,
    label="Aligned Source"
)

ax.set_title("3D Overlay: Colored by Ground-Truth Label")
ax.set_xticks([]); ax.set_yticks([]); ax.set_zticks([])

# Flatten z-axis so the layers look close
ax.set_zlim(0, 0.03)
ax.set_box_aspect([1, 1, 0.05])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

# Target (true)
plt.scatter(
    target_coordinates[:,0],
    target_coordinates[:,1],
    c=colors_tar,
    s=8,
    alpha=0.40,
    label="Target"
)

# Aligned source (predicted)
plt.scatter(
    new_rigid_coordinates[:,0],
    new_rigid_coordinates[:,1],
    c=colors_src,
    s=8,
    alpha=0.40,
    label="Aligned Source"
)

# plt.title("2D Overlay: Ground-Truth Label Coloring", fontsize=14)
plt.axis("equal")
plt.xticks([]); plt.yticks([])
for sp in plt.gca().spines.values(): 
    sp.set_visible(False)

plt.legend(frameon=False)
plt.tight_layout()
plt.show()
```
